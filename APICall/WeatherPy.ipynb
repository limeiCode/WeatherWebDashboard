```python
%matplotlib inline       

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy.core as owm 
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500 )    
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)   
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to our cities list
    if city not in cities:  
         cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
 
units = "metric"                                         
settings = {"units": units, "appid": api_key}
summary = ["name","clouds.all","sys.country", "dt", "main.humidity","coord.lat", "coord.lon", "main.temp_max","wind.speed"] #'base','cod',"id","main.temp", "main.pressure","main.temp_min","sys","timezone","visibility","weather"]
citydata = []

indx = 0
print("Beginning Data Retrieval")     
print("-----------------------------")
for city in cities:
    st = indx // 50 + 1
    rec = indx % 50
    try: 
        NestedDict_currentdata = owm.get_current(city, **settings)    # city_weather = requests.get(city_url).json()
        items = NestedDict_currentdata(*summary)    
        citydata.append(items)
    except:
        print(f"'{city}' not found. Skipping...")
    else:
        print(f"Processing Record {rec} of Set {st} | {city}") 
        indx = indx + 1
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
citydata=pd.DataFrame(citydata)
citydata=citydata.rename(columns={0:"City",1:"Cloudiness",2:"Country",3:"Date",4:"Humidity",5:"Lat",6:"Lng",7:"Max Temp",8:"Wind Speed"})
citydata.to_csv(output_data_file, index=False)
citydata.count()
```

```python
data_df = pd.read_csv(output_data_file)
data1_df = data_df[["Lat","Max Temp"]]
data2_df = data_df[["Lat","Humidity"]]
data3_df = data_df[["Lat","Cloudiness"]]
data4_df = data_df[["Lat","Wind Speed"]]

from datetime import datetime 
dt = datetime.fromtimestamp(data_df["Date"][0]) 
date_y = str(dt.year)[2:]
date_m = dt.month
date_d = dt.day
datestr=f"{date_m}/{date_d}/{date_y}" 
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
x_limit = 80   
y_limit = 60    
y_values = data1_df["Max Temp"]
x_axis = data1_df["Lat"]     

xtick_locations = np.arange(-60,120,20)    
xtick_labels = ["-60","-40", "-20",   "0",  "20",  "40",  "60",  "80", "100","120"]  
ytick_locations = np.arange(-10,60,10)    
ytick_labels = ["-10","0","10","20","30","40","50"]   

plt.xticks(xtick_locations, xtick_labels, size=6,rotation=0 , ha="right")    
plt.yticks(ytick_locations, ytick_labels, size=6, rotation=0 , ha="right") 

plt.grid()                                                
plt.tight_layout()   
plt.xlabel("Latitude", fontsize=10)             
plt.ylabel("Max Temperature(F)", fontsize=10)

plt.scatter(x_axis, y_values, marker="o", facecolors="blue",  edgecolors="black", linewidth=1, alpha=0.25) 

plt.xlim(-60, x_limit) 
plt.ylim(-10, y_limit)  
plt.legend(loc="best", fontsize=8)  

title = "City Latitude vs. Max Temperature"
plt.title(f"{title} ({datestr})", fontsize=12)

plt.figure(figsize=(30,3),dpi=200)             

plt.savefig("LatVSMaxTemp_fig1.png")

plt.show();

plt.close();
```
